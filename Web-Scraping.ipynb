```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "http://quotes.toscrape.com/page/1/"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
#extracting the quotes 
for txt in soup.find_all('span' , class_ = 'text'):
    quote = txt.get_text()
    print(quote)
```

```python
#extracting the authors' names
for author in soup.find_all('small' , class_ = 'author'):
    authors = author.get_text()
    print(authors)
```

```python
#extracting the links in website and the text displaying for links
for article in soup.find_all('a'):
    text = article.text
    href = article.get('href')
    print("text : " , text , "link : " , href)
```

```python
#extracting the tangs and links associated with them
for ahref in soup.find_all('a' , class_ = 'tag'):
    text = ahref.text
    href = ahref.get('href')
    print("tags :" , text , "link : " , href)
```

```python
#extracting the tags
for ahref in soup.find_all('a' , class_ = 'tag'):
    text = ahref.text
    print(text)
```

```python
#looping though first four pages
page = 1
while page != 5:
    url = f"http://quotes.toscrape.com/page={page}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    for txt in soup.find_all('span' , class_ = 'text'):
        quote = txt.get_text()
        print(quote)
    for author in soup.find_all('small' , class_ = 'author'):
        authors = author.get_text()
        print(authors)
    for ahref in soup.find_all('a' , class_ = 'tag'):
        text = ahref.text
        print(text)
        page = page + 1
```
